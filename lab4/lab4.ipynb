```python
import pandas as pd
import numpy as np

wizards = pd.read_excel('lab.xlsx', sheetname="1")
magicians = pd.read_excel('lab.xlsx', sheetname="2")

def constructArray(array, units):
    for index, unit in units.iterrows():
        if index > 0:
            array = np.append(array,
                [[unit['Magic Efficiency'],
                unit['Ph. Att. Efficiency'],
                unit['Magic Restore'],
                unit['Ph. Att. Restore']]], 
                axis = 0)
    return array

foe_army = np.array([[
        wizards['Magic Efficiency'][0], 
        wizards['Ph. Att. Efficiency'][0], 
        wizards['Magic Restore'][0],
        wizards['Ph. Att. Restore'][0]
]])
foe_army = constructArray(foe_army, wizards)

for index, wizard in wizards.iterrows():
    print (wizard['Wizards'] + ': ' + str(foe_army[index]))

my_army = np.array([[
    magicians['Magic Efficiency'][0],
    magicians['Ph. Att. Efficiency'][0],
    magicians['Magic Restore'][0],
    magicians['Ph. Att. Restore'][0]
]])
my_army = constructArray(my_army, magicians)

print
count = 0
for col in magicians.columns:
    if col != "id" and col != "Magicians" and col != "Price":
        print col + ": " + str(my_army.T[count])
        count += 1

def maxOfProduct(x, y):
    z = []
    for x1 in x:
        for y1 in y.T:
            z.append(max(np.multiply(x1, y1)))
    return np.array(z).reshape((x.shape[0], y.shape[1]))

print "\nEfficiency of skills (row - wizards, columns - magicians):\n"
max_skills_of_magicians = maxOfProduct(foe_army, my_army.T)
str_magician = "Wizards "
for x in magicians['Magicians'].values:
    str_magician += x + " "
print str_magician
for index, wizard in wizards.iterrows():
    print (wizard['Wizards'] + ': ' + str(max_skills_of_magicians[index]))
    
print "\nEfficiency of skills (row - magicians, columns - wizards):\n"
max_skills_of_wizards = maxOfProduct(my_army, foe_army.T)
str_wizard = "Magicians "
for x in wizards['Wizards'].values:
    str_wizard += x + " "
print str_wizard
for index, magician in magicians.iterrows():
    print (magician['Magicians'] + ': ' + str(max_skills_of_wizards[index]))

print "\n1 задание: собрать максимально эффективную армию против колдунов за минимальное количество золота"
print "\nМоя армия против:\n"
print "Костяного Дракона (кол-во: 1) выстоет Титан (кол-во: 1)"
print "Ведьмы (кол-во: 2) выстоят Феи (кол-во: 2)"
print "Вампиры (кол-во: 3) выстоят Рыцари (кол-во: 3)"
print "Зомби (кол-во: 5) выстоят Крестьяне (кол-во: 5)"
print "Скелетов (кол-во: 8) выстоят Крестьяне (кол-во: 8)"

print "\nИтого: " + str(390 + 2*200 + 3*190 + 5*90 + 8*90) + " золота"

```
